Subset and slicing:

```python
#Return name column
df["name"]
#Return multiple columns
df[["name","class","mark"]]
#Method can be applied
df[["name","class","mark"]].head()
#Mean of marks
df.mark.mean()
#Returns mark value of index 0
df.mark[0]
```

loc() and iloc()

```python
#Returns all rows in name column
df.loc[:, "name"]
#Returns all rows in name column from index 10
df.loc[10:, "name"]
#Returns first 5 rows of name and class columns
df.loc[0:4,["name","class"]]
```

```python
#1 is index for name column
#2 is index for class column
#stopping range is included in output
df.iloc[0:5,[1,2]]
#First 5 Rows and 2 columns
df.iloc[0:5,0:2]
```

Filter:

```python
#Returns Boolean
df["mark"] > 90
#Returns data
df[df["mark"]>90]
#Multiple conditions with AND operator
df[(df["class"]=="Six") & (df["gender"] == "male")]
#Multiple conditions with OR operator
df[(df["gender"] == "female") | (df["mark"] >= 90)]
#Using Loc to filter
df.loc[df["gender"]=="male",["name","class"]]
```

```python
#Returns all data where mark > 78
df.query("mark > 78")
#Returns all data where mark = 78 and gender = male
df.query("(mark > 78) and (gender == 'male')")
```

```python
df.mark.mean()
```

```python
avg_male = df.query("gender == 'male'")["mark"].mean()
avg_female = df.query("gender == 'female'")["mark"].mean()
print("Male:",avg_male,"Female:",avg_female)
```

```python
avg_male = df.loc[df["gender"]=="male", "mark"].mean()
avg_female = df.loc[df["gender"]=="female", "mark"].mean()
print("Male:",avg_male,"Female:",avg_female)
```

```python
low = df.mark.min()
high = df.mark.max()
print("Lowest mark:",low,"Highest mark:",high)
```

```python
male = df.query("gender == 'male'")["mark"].max()
female = df.query("gender == 'female'")["mark"].max()
print("Male highest:",male,"Female highest:",female)
```

```python
df.loc[df["class"]=="Five","mark"].mean()
```

```python
df.loc[(df["class"]=="Six") & (df["gender"]=="male"),"mark"].mean()
```

Sorting:

```python
#Sorts by marks
df.sort_values(by  ="mark")
#Sort by marks in Descending order
df.sort_values(by  ="mark", ascending=False)
#Sort values where marks > 50 sorted by class
df.query("mark >50").sort_values(by="class")
```
